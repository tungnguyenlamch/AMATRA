# Google Translator Evaluation on OpenMantra

This notebook evaluates the Google Translator on the OpenMantra manga translation dataset.

```python
import os
import sys

ENDWITHS = 'OpenMantra'

NOTEBOOK_DIR = os.getcwd()

if not NOTEBOOK_DIR.endswith(ENDWITHS):
    raise ValueError(f"Not in correct dir, expect end with {ENDWITHS}, but got {NOTEBOOK_DIR} instead")

BASE_DIR = os.path.abspath(os.path.join(NOTEBOOK_DIR, '..', '..', '..', '..', '..'))
print(BASE_DIR)

sys.path.insert(0, os.path.join(BASE_DIR, 'src'))
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..', '..')))
```

```python
from OpenMantraEvaluator import OpenMantraEvaluator
from pipeline.TranslationModels.GoogleTranslator import GoogleTranslator
```

```python
from dotenv import load_dotenv
load_dotenv(os.path.abspath(os.path.join(BASE_DIR, '..', '.env')))

HF_TOKEN = os.getenv('HF_TOKEN')
```

```python
OPENMANTRA_ROOT = os.path.join(BASE_DIR, 'data', 'open-mantra-dataset')
print(f"OpenMantra dataset root: {OPENMANTRA_ROOT}")

SAVE_DIR = os.path.join(BASE_DIR, 'output', 'pipeline', 'Evaluators', 'Translators', 'OpenMantra')
```

```python
# Create evaluator
evaluator = OpenMantraEvaluator(
    openmantra_root=OPENMANTRA_ROOT,
    source_lang='text_ja',
    target_lang='text_en'
)

# Create Google Translator
# use_official_api=False uses the free googletrans library
# use_official_api=True requires Google Cloud credentials
translation_model = GoogleTranslator(
    source_lang='ja',
    target_lang='en',
    use_official_api=False,  # Set to True if you have Google Cloud credentials
    verbose=True
)
```

```python
# Quick test to verify the translator works
translation_model.load_model()

test_texts = [
    "こんにちは",
    "お前はもう死んでいる",
    "なに！？"
]

results = translation_model.predict(test_texts)
for src, tgt in zip(test_texts, results):
    print(f"{src} -> {tgt}")

translation_model.unload_model()
```

```python
# Run full evaluation
metrics = evaluator.evaluate(translation_model, save_dir=SAVE_DIR)
```

```python
# Display metrics
print("\n=== Evaluation Results ===")
for key, value in metrics.items():
    if isinstance(value, float):
        print(f"{key}: {value:.4f}")
    else:
        print(f"{key}: {value}")
```
