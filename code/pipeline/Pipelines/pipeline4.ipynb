```python
import os
import sys
import matplotlib.pyplot as plt
import cv2

# 1. Setup Paths
NOTEBOOK_DIR = os.getcwd()
BASE_DIR = os.path.abspath(os.path.join(NOTEBOOK_DIR, '..', '..', '..'))
sys.path.insert(0, os.path.join(BASE_DIR, 'code'))

# 2. Import the new Pipeline Class
from pipeline.Utils.MangaPipeline import MangaPipeline
```

```python
# 3. Initialize Pipeline (Loads models once)
YOLO_PATH = os.path.join(BASE_DIR, "models","bubble-detection","YOLO_Ultralytics","super_yolo.pt")
pipeline = MangaPipeline(YOLO_PATH)
```

```python
# 4. Define Image Path
IMG_PATH = os.path.join(BASE_DIR, "data/Manga109_released_2023_12_07/images/Count3DeKimeteAgeru/005.jpg")
img = cv2.imread(IMG_PATH)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 5. Run the Pipeline
final_image, bubble_data = pipeline.process(img_rgb)
```

```python
# 6. Display Result
plt.figure(figsize=(15, 20))
plt.imshow(final_image)
plt.axis('off')
plt.show()
```

```python
# 7. Inspect Text Data (Optional)
for i, b in enumerate(bubble_data):
    print(f"Bubble {i}:")
    print(f"  OCR: {b['ocr_text']}")
    print(f"  Trans: {b['translated_text']}\n")
```
